### Introduction

The purpose of this analysis is to create a player recommendation engine, enabling clubs to identify a suitable player for their team and then find statistically similar players of differing age or values to identify transfer targets. 

The approach is to: 
- Find stats of a particular player by their name
- Compare their stats to all other players and calculate similarity (both Euclidean and Cosine Similarity have been considered).
- Create a function that: 
    - Takes inputs of name of a player, money to spend, age, position 
    - Identifies the top 5 similar players names, stats, market value, modelled value, similarity score
    - Generates pizza plots of each 5 players stats
    
This analysis is inclusive of: 
- Modelling: Players who are under and over valued 
- Clustering: Player positions & identified clusters

#### Function: Retrieve individual player stats

```python
# Function to retrieve individual player stats from the dataframe
def getplayerstats(name):
    stats = df_recommender[df_recommender['full_name']==name].iloc[:1,15:]
    #club = player_df_reduced[player_df_reduced['full_name']==name]['Current Club'].iloc[0]
    return stats
```

#### Function: Calculate Cosine Similarity

```python
# Write Function to compare two players - by cosine similarity
def comparetwoplayers(name1, name2):
    cosine_sim = cosine_similarity(getplayerstats(name1), getplayerstats(name2))
    return cosine_sim[0][0]
```

```python
# Example - 'Harry Kane, Erling Haaland'
comparetwoplayers('Harry Kane', 'Erling Haaland')
```

#### Function: Plot key player statistics
This function derives the following:
- Identifies player statistics based on a subset of selected important features (domain knowledge)
- Calculates percentiles of those statistics against all other players
- Plots the statistics in a 'pizza plot'

```python
# Function to plot player statistics, including the percentile of their statistics
def get_player_stats_plots(player_names):

    slice_colors = ["blue"] * 3 + ["green"] * 3 + ["red"] * 2
    text_colors = ["white"]*8
    font_normal = FontManager()
    font_bold = FontManager()
    
    for name in player_names:
        player_stats = market_value_position_final.loc[market_value_position_final["full_name"] == name]
        player_stats = player_stats.loc[player_stats['season'] == '2022/2023']
        team = list(player_stats['Current Club'])[0]
        league = list(player_stats['league'])[0]
        player_stats = player_stats[['npxg_per_90_overall', "goals_per_90_overall", "shots_on_target_per_90_overall",
                                     "assists_per_90_overall", "key_passes_per_90_overall", "dribbles_per_90_overall", 
                                     "duels_won_per_90_overall", "aerial_duels_won_per_90_overall"]]   
        
        per_90_columns = player_stats.columns[:]
        values = [round(player_stats[column].iloc[0], 2) for column in per_90_columns]
        percentiles = [int(stats.percentileofscore(df_recommender[column], player_stats[column].iloc[0])) for column in per_90_columns]
        

        # Create PyPizza plot for each player
        baker = PyPizza(
            params=per_90_columns,
            min_range=None,
            max_range=None,
            straight_line_color="#000000",
            straight_line_lw=1,
            last_circle_lw=1,
            other_circle_lw=1,
            other_circle_ls="-."
        )


        # Plot pizza for the player
        fig, ax = baker.make_pizza(
            percentiles,
            figsize=(5, 5),
            param_location=110,
            slice_colors=slice_colors,
            value_colors=text_colors,
            value_bck_colors=slice_colors,
            kwargs_slices=dict(
                facecolor="cornflowerblue", edgecolor="#000000",
                zorder=2, linewidth=1
            ),
            kwargs_params=dict(
                color="#000000", fontsize=8,
                fontproperties=font_normal.prop, va="center"
            ),
            kwargs_values=dict(
                color="#000000", fontsize=8,
                fontproperties=font_normal.prop, zorder=3,
                bbox=dict(
                    edgecolor="#000000", facecolor="cornflowerblue",
                    boxstyle="round,pad=0.2", lw=1
                )
            )
        )

        # Set values as text on the plot
        texts = baker.get_value_texts()
        for i, text in enumerate(texts):
            text.set_text(str(values[i]))

        # Add title
        fig.text(
            0.515, 0.97, f"{name} per 90 - {team}", size=12,
            ha="center", fontproperties=font_bold.prop, color="#000000"
        )

        # Add subtitle
        fig.text(
            0.515, 0.942,
            f"{league} | Season 2022-2023",
            size=8,
            ha="center", fontproperties=font_bold.prop, color="#000000"
        )

        plt.show()
    
    return 
```

#### Function: Recommendation Engine

This function takes inputs from all of the previous functions in order to generate a single table. 

The function enables teams to input: 
1. A player of interest
2. A maximum player age
3. A maximum market value they would want to pay
4. The option of entering a position - This is taken from the clustering analysis 
5. The option of entering a league - if they are interested in players only from a specific league

The output is a shortlist of players to review, including the modelling predictions of market value, whether under / over valued, and by how much. 

There is also a calculation factoring on how similar the player is to the one of interest and how under or overvalued they are to support the decision making process. 

```python
def most_similar_players_cos(name1, age, marketvalue, position = None, league = None):
    similarity_scores = []
    
    
  # Filter players based on criteria
    filtered_players = df_recommender[
        (df_recommender['age'] <= age) &
        (df_recommender['Actual Market Value (M)'] < marketvalue)
    ]
    
    if position is not None:
        filtered_players = filtered_players[filtered_players['Position_NEW'] == position]
    
    if league is not None:
        filtered_players = filtered_players[filtered_players['league'] == league]
    
    # If no players match - None
    if filtered_players.empty:
        return None
    
    all_players = filtered_players['full_name'].unique()
    
    # Calculate similarity score for player vs all other players
    for name in all_players:
        similarity_scores.append((
        name1, 
        name, 
        round(comparetwoplayers(name1, name)*100), 
        df_recommender[df_recommender['full_name'] == name]['age'].iloc[0], 
        df_recommender[df_recommender['full_name'] == name]['Current Club'].iloc[0], 
        df_recommender[df_recommender['full_name'] == name]['Actual Market Value (M)'].iloc[0], 
        df_recommender[df_recommender['full_name'] == name]['Predicted Market Value (M)'].iloc[0],
        df_recommender[df_recommender['full_name'] == name]['Prediction Difference'].iloc[0],
        df_recommender[df_recommender['full_name'] == name]['Under / Over'].iloc[0] 
    ))

    # Create DataFrame of similarity scores
    df_similarity = pd.DataFrame(similarity_scores, columns=['Player',
                                                             'Similar Player', 
                                                             'Cosine Similarity', 
                                                             'Age', 
                                                             'Current Club', 
                                                             'Actual Market Value (M)', 
                                                             'Predicted Market Value (M)',
                                                             'Prediction Difference',
                                                             'Under / Overvalued'])
    df_similarity['Similarity & Value'] = df_similarity['Cosine Similarity'] -df_similarity['Prediction Difference']
    
    df_similarity = df_similarity.sort_values('Cosine Similarity', ascending=False).head(5).reset_index(drop=True)

    display(df_similarity)
    
    # Get player statistics plots
    player_stats_plots = get_player_stats_plots([name1] + list(df_similarity['Similar Player']))
      
    return player_stats_plots
```

### Example with Lionel Messi

```python
# Players similar to lionel messi from any league and any position
most_similar_players_cos('Marcus Rashford', 30, 100, league = None, position = None)
```

#### Insights

Based on domain knowlege, it is evident these players are similar in playing styles and also statistically similar. 

It is evident from the pizza plots how the similarity has been considered, with dribbles, duels won and key passess all coming out as in the top percentile. 

###Â Euclidean Distance

I also looked into Euclidean distance as a different measure of similarity, here there is more emphasis on the scale of the values, rather than just the direction. 

```python
from scipy.spatial.distance import euclidean
```

```python
# Function to retrieve individual player stats from the dataframe
def getplayerstats_euc(name):
    stats = np.array(df_recommender[df_recommender['full_name']==name].iloc[:1,13:]).tolist()[0]
    return stats
```

```python
def comparetwoplayers_euc(name1, name2):
    euclidean_distance = euclidean(getplayerstats_euc(name1), getplayerstats_euc(name2))
    return euclidean_distance
```

```python
def euclidean_shortlist(name1, position, age, marketvalue):
    euclidean_scores = []
    
     # All player names
    all_players = df_recommender[
    (df_recommender['Position_NEW'] == position) &
    (df_recommender['age'] <= age) &
    (df_recommender['Actual Market Value (M)'] < marketvalue)
    ]['full_name'].unique()
    
    # Get similarity score for player vs all other players
    for name in all_players:
        euclidean_scores.append((name1,name,comparetwoplayers_euc(name1, name)))
    
    # Final Data Frame
    df1 = pd.DataFrame(euclidean_scores, columns = ['Player', 'Similar Player', 'Euclidean Distance'])\
    .sort_values('Euclidean Distance', ascending = True).head(5).reset_index()
    
    #Â List of similar players
    list_of_similar_players = list(df1['Similar Player'])
    
    return print(f'{df1}\n\n'),get_player_stats_plots([name1]), get_player_stats_plots(list_of_similar_players)
```

### Example with Lionel Messi

```python
euclidean_shortlist('Lionel Messi', 'Strikers', 30, 100)
```

####Â Insights

There are similiar players in the list to the ones for Lionel Messi utilising cosine similarity. 

For the purpose of this engine I will utilise Cosine similarity. In the context of what teams are looking for, it would be preferable to find a player who plays in a similar way but could potentially be cheaper so this will be taken forward for the final product.
