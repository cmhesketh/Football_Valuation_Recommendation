## Functions for Data Loading and Data Checks

```python
# Check all rows and columns in CSV file and return number of rows, file names, number of columns, no. files
def check_rows_columns_in_csv_files(folder_path):
    csv_files = [file for file in os.listdir(folder_path) if file.endswith('.csv')]
    file_info=[]
    number_of_files = len(file_info)
    total_rows = 0
    first_columns = None
    for file in csv_files:
        file_path = os.path.join(folder_path, file)
        df = pd.read_csv(file_path)
        total_rows += df.shape[0]
        file_info.append((file_path, df.shape))
        if first_columns is None:
            first_columns = df.columns.tolist()  
        else:
            if df.columns.tolist() != first_columns:  
                return False, total_rows, len(file_info), file_info
    return True, total_rows, len(first_columns), len(file_info), file_info
```

```python
# Take files from folder path and concatenate them
def concatenate_csv_files(folder_path):
    csv_files = [file for file in os.listdir(folder_path) if file.endswith('.csv')]
    dfs = []
    for file in csv_files:
        file_path = os.path.join(folder_path, file)
        df = pd.read_csv(file_path)
        df['file_name'] = file
        dfs.append(df)
    return pd.concat(dfs, ignore_index=True)
```

```python
# Check unique values in each column of dataframe
def get_unique_values(df):
    col_unique = []
    for i in df.columns:
        unique_values = df[i].unique()
        col_unique.append((i,unique_values))
    return col_unique
```

```python
# Check for special characters in files
def check_special_characters(df):
    special_char_location= []
    for column in df.columns:
        for value in df[column].items():
            if isinstance(value, str):
                matches = bool(re.search(r'[^\w\s\-\.\/]', value))
                for match in matches: 
                    special_char_location.append((index, column))
    return special_char_location
```

```python
# Locate the special characters
def locate_special_characters(df):
    special_chars_locations = []
    for column in df.columns:
        for index, value in df[column].items():
            if isinstance(value, str):
                matches = re.finditer(r'[^\w\s\-\.\/\'\&]', value)
                for match in matches:
                    special_chars_locations.append((index, column, value, match.start()))
    return special_chars_locations
```

#### Challenges with merging based on player name

The challenge with merging the two datasets together is that players have different names, spellings, special characters, shorthand names in both data sets. 

As a result, fuzzy matching has been utilised to simplify the mapping key across the two tables. 

In order to do support the matching, all names were converted to lower case and unidecode was used to remove special characters in advance. 

```python
# Adjust name to be only lowercase, with a '-', and decode special characters
player_df_drop_na['map_name']=player_df_drop_na['full_name'].str.lower().str.replace(' ', '-')
player_df_drop_na['map_name'] = player_df_drop_na['map_name'].apply(unidecode)
```

```python
# Get unique player names from each table
player_name_valuations = valuations['player_code'].unique().tolist() 
player_name_players= player_df_drop_na['map_name'].unique().tolist() 
```

#### Fuzzy Matching

Utilise fuzzy matching to generate mapping player names across the two tables. This generates a set of matching scores to assess. It is not a perfect science but manual review also took place to ensure a high level of accuracy. 

```python
# Generate matches using fuzzy match
matches_valuations = [process.extractOne(name, player_name_valuations)[0:2] for name in player_name_players]
```

```python
# Exampple of matches
matches_valuations[0:5]
```

```python
# 80% are correct first time - After review, i will cut off at a 90% match
pd.DataFrame(matches_valuations).value_counts(1).sort_values(ascending = False)
```

#### Bringing it together
- Create a dataframe to compare the actual player name in the player file, to the matched name in the valuation file and assess the scoring.

- Create a mapping table across the two dataframes

- Merge the dataframes to bring the market value into the dataframe holding detailed player statistics

- Take an average of rows where a player has played for multiple teams within a single season. This will average their statistics for example if they have been on loan once or twice in a single year. 

```python
# Pull out matches & Scores from the datafrane
name_match_valuations =[match[0] for match in matches_valuations] 
name_score_valuations =[match[1] for match in matches_valuations]
```

```python
# Show in a data frame
name_matches = pd.DataFrame({'player_name': player_name_players, 'match':name_match_valuations, 'score':name_score_valuations})

# Generate list of names to keep 
list_of_names = list(name_matches[name_matches['score']>90].iloc[:,0])
```

```python
# Comparison of player names from the player file and the match from the valuation file
name_matches.head()
```

```python
# Create mapping table to support keys 
players_to_map = name_matches[name_matches['score']>90]
player_table = list(players_to_map.iloc[:,0])
valuation_table = list(players_to_map.iloc[:,1])
mapping_table = dict(zip(player_table, valuation_table))
```

```python
# Map names (we expect NAs to be present now)
player_df_drop_na['mapping_name'] = player_df_drop_na['map_name'].map(mapping_table)
```

```python
#¬†Drop rows with NA in the mapping_name column
full_player_table = player_df_drop_na.dropna(subset=['mapping_name'])
```

```python
# Map the Market Value from the transfers table into this table 
#¬†Create mapping key
full_player_table['player_year_key'] = full_player_table['mapping_name']+'-'+full_player_table['season']
```

```python
# Create the merge
player_value = pd.merge(full_player_table, valuations[['mapping_key', 'market_value_in_eur']], 
                        left_on = 'player_year_key', right_on = 'mapping_key', how = 'left')
```

```python
# Check for NA values
player_value.isna().sum()
```

```python
# Sorted by market value
player_value.sort_values('market_value_in_eur', ascending = False).T.head(5)
```

```python
# Drop columns no longer needed
player_value1 = player_value.drop(columns = ['map_name', 'mapping_name', 'mapping_key'])
```

```python
# Take an average where there are multiple records per year for each player
player_value_final = player_value1.groupby(['full_name', 'age', 'birthday', 'birthday_GMT', 'league', 
                      'season', 'position', 'Current Club', 'nationality', 'file_name', 'player_year_key']).mean().reset_index()
```

```python
# Drop columns where market value is null
player_value_final1 = player_value_final.dropna(subset=['market_value_in_eur'])
```
